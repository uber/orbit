# Diagnostic Visualization demo

In this demo, we want to demonstrate how to use the plotting utilities to visualize the posterior samples in some of the Orbit models (e.g. **SVI** and **Full Bayesian**) . Those could be very useful in practice when checking the convergence status of the model.

```python
import pandas as pd
import numpy as np

from orbit.models import LGT, DLT
from orbit.diagnostics.plot import plot_posterior_params
from orbit.utils.dataset import load_iclaims

import matplotlib.pyplot as plt

%matplotlib inline
```

```python
%load_ext autoreload
%autoreload 2
```

## load data

```python
raw_df = load_iclaims()

raw_df.dtypes
```

```python
raw_df.head()
```

```python
df=raw_df.copy()

test_size=52

train_df=df[:-test_size]
test_df=df[-test_size:]
```

## Fit a Model

```python
DATE_COL="week"
RESPONSE_COL="claims"
REGRESSOR_COL=['trend.unemploy', 'trend.filling', 'trend.job']
```

```python
dlt_mcmc = DLT(response_col=RESPONSE_COL,
               date_col=DATE_COL,
               regressor_col=REGRESSOR_COL,
               regressor_sign=["+", '+', '='],
               seasonality=52,         
               num_warmup=2000,
               num_sample=2000,
               chains=4)
```

## Posterior Diagnostic Viz

```python
dlt_mcmc.fit(df=train_df)
```
